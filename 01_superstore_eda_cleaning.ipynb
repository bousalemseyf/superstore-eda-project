## ðŸ§¹ Data Cleaning Steps

This section explains how the raw Superstore dataset was cleaned:
- Checked the first rows and data types.
- Verified missing values.
- Converted date columns to datetime format.
- Created `Year`, `Month`, and `Quarter` columns for easier time-based analysis.
- Saved the cleaned dataset for further use.


```python
import pandas as pd

# 1ï¸âƒ£ Load your raw CSV
df = pd.read_csv('../data/Superstore.csv', encoding='ISO-8859-1')

# 2ï¸âƒ£ Quick checks
print("âœ… FIRST 5 ROWS:")
print(df.head())

print("\nâœ… INFO:")
print(df.info())

print("\nâœ… NULL VALUES:")
print(df.isnull().sum())

# 3ï¸âƒ£ Convert 'Order Date' and 'Ship Date' to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# 4ï¸âƒ£ Add Year, Month, Quarter columns
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Quarter'] = df['Order Date'].dt.quarter

# 5ï¸âƒ£ Save cleaned version
df.to_csv('../data/Superstore_Cleaned.csv', index=False)

print("\nâœ… DONE! Cleaned file saved as 'Superstore_Cleaned.csv' in your 'data/' folder.")
```

```python
## Total Sales and Profit

This step calculates the overall total sales and profit for the Superstore dataset.  
It shows how much revenue and profit the business has generated in total.

```

```python
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()

print(f"ðŸ’° Total Sales: ${total_sales:,.2f}")
print(f"ðŸ’¸ Total Profit: ${total_profit:,.2f}")
```

```python
## Sales and Profit by Year

This shows how sales and profit have changed each year.
It helps identify trends and growth over time.
```

```python
sales_by_year = df.groupby('Year')[['Sales', 'Profit']].sum().reset_index()
print(sales_by_year)
```

```python
## Top 10 States by Profit

This step finds the top 10 U.S. states that contribute the most to overall profit.
It helps identify the most profitable regions for the business.
```

```python
top_states = df.groupby('State')['Profit'].sum().sort_values(ascending=False).head(10)
print(top_states)
```

```python
## Average Discount by Region

This step calculates the average discount given in each region.
It helps understand how discounting strategies vary by region.
```

```python
avg_discount = df.groupby('Region')['Discount'].mean().reset_index()
print(avg_discount)
```

### Insights: Average Discount by Region
- Central region has the highest average discount.
- West region has the lowest, which may affect competitive pricing.


```python
## Profit by Category and Sub-Category

This shows profit totals by product category and sub-category.
It helps identify which product lines are most profitable.
```

```python
profit_by_category = df.groupby(['Category', 'Sub-Category'])['Profit'].sum().sort_values(ascending=False)
print(profit_by_category)
```

### Insights: Profit by Category & Sub-Category
- Technology category, especially Copiers & Phones, are top profit generators.
- Some sub-categories like Tables show negative profit, indicating potential issues.


```python
## Visualize Sales and Profit Trends

This creates simple bar charts to visualize sales and profit by year and the top states by profit.
```

### ðŸ“Š Insights: Sales & Profit by Year

```python
# Top 10 States by Profit
df.groupby('State')['Profit'].sum().sort_values(ascending=False).head(10).plot(kind='bar', figsize=(8,6))
plt.title('Top 10 States by Profit')
plt.xlabel('State')
plt.ylabel('Profit ($)')
plt.show()
```

### ðŸ—ºï¸ Insights: Top 10 States by Profit

## ðŸ“Š Customer Segmentation: Sales & Profit by Segment

This section analyzes sales and profit for each customer segment (`Consumer`, `Corporate`, `Home Office`).  
It helps identify which segments drive the most revenue and profit.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group by Segment
segmentation = df.groupby('Segment')[['Sales', 'Profit']].sum().reset_index()

# Plot with Seaborn
plt.figure(figsize=(8,5))
segmentation_melt = segmentation.melt(id_vars='Segment', var_name='Metric', value_name='Amount')

sns.barplot(data=segmentation_melt, x='Segment', y='Amount', hue='Metric')
plt.title('Sales & Profit by Customer Segment')
plt.ylabel('Amount ($)')
plt.show()
```


âœ… **Insights:**
- The **Consumer** segment generates the highest sales.
- However, check if profit aligns â€” sometimes profit margins vary!
- Consider strategies to boost underperforming segments.


## ðŸ“Š Sales & Profit by Product Category

This section visualizes total Sales and Profit by each product Category.
It helps identify which categories drive revenue and profit the most.


```python
# ðŸ“Š Group by Category
category = df.groupby('Category')[['Sales', 'Profit']].sum().reset_index()

# ðŸ“Œ Prepare for Seaborn
import seaborn as sns
import matplotlib.pyplot as plt

# Melt to long format
category_melt = category.melt(id_vars='Category', var_name='Metric', value_name='Amount')

# ðŸ“Š Plot with Seaborn
plt.figure(figsize=(8,5))
sns.barplot(data=category_melt, x='Category', y='Amount', hue='Metric')
plt.title('Sales & Profit by Product Category')
plt.xlabel('Category')
plt.ylabel('Amount ($)')
plt.show()
```

âœ… **Insights:**
- Technology shows the highest sales â€” does profit match?
- Watch for categories with high sales but low profit.
- Focus marketing on the most profitable product lines.


## ðŸ“Š Profit by Category & Sub-Category

This heatmap shows which product sub-categories are the most and least profitable.
It helps spot areas to focus on or improve.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create pivot table for heatmap
pivot = df.pivot_table(index='Category', columns='Sub-Category', values='Profit', aggfunc='sum', fill_value=0)

plt.figure(figsize=(12, 6))
sns.heatmap(pivot, annot=True, fmt=".0f", cmap="YlGnBu", linewidths=.5)
plt.title('Profit by Category & Sub-Category')
plt.ylabel('Category')
plt.xlabel('Sub-Category')
plt.show()
```

âœ… **Insights:**
- Quickly see which sub-categories generate the highest profit.
- Identify sub-categories with negative profit that may need review.
- This can guide pricing, promotions, or product strategy.


## ðŸ“ˆ Sales & Profit Trend Over Time

This line plot shows how Sales & Profit change each month.
It helps identify seasonality, growth patterns, or unusual spikes/drops.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group by Year & Month
monthly_trend = df.groupby(['Year', 'Month'])[['Sales', 'Profit']].sum().reset_index()

# Create a 'Date' column for plotting
import pandas as pd
monthly_trend['Date'] = pd.to_datetime(monthly_trend['Year'].astype(str) + '-' + monthly_trend['Month'].astype(str))

# Melt for Seaborn
monthly_melt = monthly_trend.melt(id_vars='Date', value_vars=['Sales', 'Profit'], var_name='Metric', value_name='Amount')

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_melt, x='Date', y='Amount', hue='Metric', marker='o')
plt.title('Monthly Sales & Profit Trend')
plt.xlabel('Date')
plt.ylabel('Amount ($)')
plt.xticks(rotation=45)
plt.show()
```

âœ… **Insights:**
- Spot seasonal peaks (e.g., end-of-year holiday spikes).
- Check if profit follows the same trend as sales.
- Useful for inventory, promotions, and forecasting.
